```python
import cv2
import torch
import numpy as np
from ultralytics import YOLO  # Import YOLO class from ultralytics

# Load the YOLOv8 model
model = YOLO('G:/runs/detect/yolov8n_person_scratch/weights/best.pt')  # Replace with your local model path

# Initialize video capture (change 'video.mp4' to your video file path)
cap = cv2.VideoCapture('C:/Users/HP/Downloads/test_vid.mp4')

# Get video properties (frame width, height, and frames per second)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define codec and create VideoWriter object to save the output video
output_path = 'C:/Users/HP/Downloads/output_video.mp4'  # Specify the output path
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for mp4
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Define a color for the bounding box and tracking line (white)
bbox_color = (255, 255, 255)
line_color = (255, 255, 255)

# Function to draw bounding box and track the center point
def draw_bbox_and_track(frame, bbox, track_points):
    x1, y1, x2, y2 = map(int, bbox)
    # Draw the bounding box
    cv2.rectangle(frame, (x1, y1), (x2, y2), bbox_color, 2)
    
    # Calculate the center of the bounding box
    center_x, center_y = (x1 + x2) // 2, (y1 + y2) // 2
    
    # Add the center point to the track points list
    track_points.append((center_x, center_y))
    

# Store points to track the center
track_points = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run the frame through the YOLOv8 model
    results = model(frame)

    # Extract the bounding boxes and confidence scores
    for result in results:
        for bbox in result.boxes.xyxy:  # Get bounding box coordinates
            x1, y1, x2, y2 = map(int, bbox[:4])  # First four values are the bounding box coordinates

            # Check if the detected object is a person (assuming class 0 is for 'person')
            if int(result.boxes.cls[0]) == 0:  # Adjust based on your model's class mapping for 'person'
                draw_bbox_and_track(frame, [x1, y1, x2, y2], track_points)

    # Write the frame with bounding box and tracking line to the output video
    out.write(frame)

    # Display the frame
    cv2.imshow('Tracking', frame)

    # Press 'q' to exit the video
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture, writer, and close windows
cap.release()
out.release()  # Save the video
cv2.destroyAllWindows()

print(f"Video saved at: {output_path}")
```
